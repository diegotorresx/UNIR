<span style="font-size: 14pt; font-weight: bold; color: #0098cd;">Pregunta 2.</span>
<span style="font-size: 14pt; font-weight: bold; color: #0098cd;">¿Cuántas palabras totales hay en los comentarios del corpus?</span>

```python
def contarPalabras(text):
    doc = nlp(str(text))
    return sum(1 for token in doc if not token.is_space and not token.is_punct)

data["num_palabras"] = data["CONTENIDO A ANALIZAR"].apply(contarPalabras)

total_palabras = data["num_palabras"].sum()
print("Número total de palabras:", total_palabras)
```

<span style="font-size: 14pt; font-weight: bold; color: #0098cd;">Característica adicional 1.</span>


<span style="font-size: 14pt; font-weight: bold; color: #0098cd;">[Proporciond e mayúsculas]</span>

```python
def proporcion_mayusculas(texto):
    palabras = str(texto).split()
    if len(palabras) == 0:
        return 0
    mayusculas = sum(1 for p in palabras if p.isupper() and len(p) > 1)
    return mayusculas / len(palabras)

data["prop_mayusculas"] = data["CONTENIDO A ANALIZAR"].apply(proporcion_mayusculas)

print(data["prop_mayusculas"].describe())
print(data.groupby("odio_binario")["prop_mayusculas"].mean())
```

<b>Las mayúsculas dentro de la redaccion digital se entiende como el hacer un enfasis, agresividad, contundencia, etc. Por esta razon considero que es necesaria para analizar este tipo de mensajes de odio y no odio

Para esto lo que hice fué contar las palabras que están en mayusculas y dividir ese resultado por la cantidad total de palabras de cada comentario.

Con el resultado obtenido pude determinar que no es una varable significativa para el estudio puesto que los valores fueron minimos por lo que su presencia para sacar conclusiones acertadas puede ser descartada.

<span style="font-size: 14pt; font-weight: bold; color: #0098cd;">Característica adicional 2.</span>


<span style="font-size: 14pt; font-weight: bold; color: #0098cd;">[Cantidad de signos de exclamación o interrogación]</span>

```python
def contar_signos_expresivos(texto):
    texto = str(texto)
    return texto.count("!") + texto.count("¡") + texto.count("?") + texto.count("¿")

data["signos_expresivos"] = data["CONTENIDO A ANALIZAR"].apply(contar_signos_expresivos)

print(data["signos_expresivos"].describe())
print(data.groupby("odio_binario")["signos_expresivos"].mean())
```

<span style="font-size: 14pt; font-weight: bold; color: #0098cd;">Característica adicional 3.</span>


<span style="font-size: 14pt; font-weight: bold; color: #0098cd;">[Longitud de las palabras]</span>

```python
def longitud_media_palabras(texto):
    doc = nlp(str(texto))
    palabras = [token.text for token in doc if not token.is_space and not token.is_punct]
    if len(palabras) == 0:
        return 0
    return sum(len(p) for p in palabras) / len(palabras)

data["long_media_palabra"] = data["CONTENIDO A ANALIZAR"].apply(longitud_media_palabras)

print(data["long_media_palabra"].describe())
print(data.groupby("odio_binario")["long_media_palabra"].mean())
```

<b>

La longitud puede dar indicios del tipo de lenguaje usado en los comentarios. Para esto se utilizo nuevamente la tokenicación que ofrece la libreria de spacy, nuevamente se eliminaron los signos de puntuación, espacios, etc. Se procedió a calcular el numero de caracteres de cada palabra. Lo que espero es que esto ayude a diferenciar mensajes sencillos a mensajes con mayor nivel de redacción.

<span style="font-size: 14pt; font-weight: bold; color: #0098cd;">Característica adicional 4.</span>


<span style="font-size: 14pt; font-weight: bold; color: #0098cd;">[Diversidad de palabras]</span>

```python
def diversidad_lexica(texto):
    doc = nlp(str(texto))
    palabras = [token.text.lower() for token in doc if not token.is_space and not token.is_punct]
    if len(palabras) == 0:
        return 0
    return len(set(palabras)) / len(palabras)

data["diversidad_lexica"] = data["CONTENIDO A ANALIZAR"].apply(diversidad_lexica)

print(data["diversidad_lexica"].describe())
print(data.groupby("odio_binario")["diversidad_lexica"].mean())
```

<b>

Propongo esta característica para determinar si los comentarios analizados contienen variedad en palabras o por el contrario son comentarios más bien uniformes repetitivos ya que en los mensajes de odio en particular las expresiones tienden a repetirse.

Para esto conte las palabrás únicas entre el número total de palabras esto después de normalizar el texto.

Con los resultados obtenidos podemos estar ante una variable que nos puede dar indicios para profundizar en el estudio de la variedad léxica usada en este tipo de mensajes.
